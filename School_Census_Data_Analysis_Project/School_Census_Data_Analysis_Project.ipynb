  - **a)** Unidad de estudio 🧑‍🎓

### 5.2. Primer Paso: Preparación y Exploración de Datos 🐍📊

- **Inspecciona:**
  - Las primeras filas del dataset 📝

### a) Unidad de estudio 🧑‍🎓🔪  
La **unidad de estudio** es la entidad individual sobre la que se recogen los datos. En este cuestionario, cada estudiante (sí, tú, que quizá prefieras quedarte dormido en clase 😴📚) es una unidad de estudio. Es como si cada alma perdida fuera observada en detalle, revelando sus oscuros secretos. ¡Bienvenidos al infierno de la información! 🔥👻

---

6. **Histograma: Tiempo de Trabajo Remunerado**  
   Construye un histograma con la variable “tiempo que trabaja con pago el estudiante en la semana (horas)” y responde:
   - **a)** ¿Qué porcentaje de los estudiantes trabajan máximo 13 horas a la semana con pago? ⏱️🔝
   - **b)** ¿Qué porcentaje de los estudiantes trabajan mínimo 3 horas a la semana con pago? ⏱️🔻
   - **c)** ¿El 20% de los estudiantes trabajan mínimo cuántas horas con pago? ⏳
   - **d)** ¿El 20% de los estudiantes trabajan máximo cuántas horas con pago? ⏳
   - **e)** ¿En qué intervalo se halla la moda? 📊
   - **f)** ¿El 50% de los estudiantes trabajan máximo cuántas horas con pago? 🎯
   - **g)** ¿En promedio, cuántas horas trabajan con pago a la semana? ¿Y cuál es la desviación estándar? 📈📉
   - **h)** ¿Qué tipo de asimetría presenta esta variable? 🔄

```python
df['Paid_Work_Hours'] = pd.to_numeric(df['Paid_Work_Hours'], errors='coerce')
df = df[(df['Paid_Work_Hours'] >= 0) & (df['Paid_Work_Hours'] <= 40)]
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(df['Paid_Work_Hours'], bins=10, color='teal', edgecolor='black')
plt.title("Histograma: Tiempo de Trabajo Remunerado (horas)")
plt.xlabel("Horas de Trabajo")
plt.ylabel("Número de Estudiantes")
plt.tight_layout()
plt.show()

# a) Porcentaje de estudiantes que trabajan máximo 13 horas
perc_max_13 = (df['Paid_Work_Hours'] <= 13).mean() * 100

# b) Porcentaje de estudiantes que trabajan mínimo 3 horas
perc_min_3 = (df['Paid_Work_Hours'] >= 3).mean() * 100

# c) El 20% superior: valor a partir del cual el 20% de los estudiantes trabajan más horas
min_hours_top20 = df['Paid_Work_Hours'].quantile(0.80)

# d) El 20% inferior: valor por debajo del cual se encuentra el 20% de los estudiantes
max_hours_bottom20 = df['Paid_Work_Hours'].quantile(0.20)

# e) Intervalo donde se halla la moda: se selecciona el bin con mayor frecuencia
max_bin_index = np.argmax(counts)
mode_interval = (bins[max_bin_index], bins[max_bin_index + 1])

# f) La mediana (50% de los estudiantes trabajan hasta esa cantidad de horas)
median_hours = df['Paid_Work_Hours'].median()

# g) Promedio y desviación estándar
mean_hours = df['Paid_Work_Hours'].mean()
std_hours = df['Paid_Work_Hours'].std()

# h) Asimetría: calcular el coeficiente de skewness
skewness = df['Paid_Work_Hours'].skew()
if skewness > 0:
    skew_type = "asimetría positiva (sesgada a la derecha)"
elif skewness < 0:
    skew_type = "asimetría negativa (sesgada a la izquierda)"
else:
    skew_type = "simétrica"

print(f"a) Porcentaje que trabajan máximo 13 horas: {perc_max_13:.2f}%")
print(f"b) Porcentaje que trabajan mínimo 3 horas: {perc_min_3:.2f}%")
print(f"c) El 20% superior trabaja al menos: {min_hours_top20:.2f} horas")
print(f"d) El 20% inferior trabaja como máximo: {max_hours_bottom20:.2f} horas")
print(f"e) La moda se encuentra en el intervalo: {mode_interval}")
print(f"f) La mediana (50% de los estudiantes) trabaja como máximo: {median_hours:.2f} horas")
print(f"g) Promedio: {mean_hours:.2f} horas, Desviación estándar: {std_hours:.2f}")
print(f"h) Tipo de asimetría: {skew_type} (skewness = {skewness:.2f})")
```

9. **Tabla Cruzada: Horas de Sueño vs. Género**  
   Construye una tabla cruzada entre la variable “¿Cuántas horas de sueño por noche suele dormir cuando no tiene clases al día siguiente?” y “Género”. Responde:
   - **a)** ¿Qué porcentaje de estudiantes de la muestra son mujeres?
   - **b)** ¿Cuántas horas de sueño en promedio tienen los estudiantes? 😴
   - **c)** ¿En términos generales, qué género duerme más tiempo?
   - **d)** ¿Qué género tiene tiempos de sueño más homogéneos?
   - **e)** ¿Qué género presenta una mayor simetría en los tiempos de sueño?
   - **f)** ¿Qué género tiene un histograma más puntiagudo? ¿Y cuál uno más aplanado? 📉📊

```python
df['Sleep_Hours_Non_Schoolnight'] = pd.to_numeric(df['Sleep_Hours_Non_Schoolnight'], errors='coerce')
crosstab = pd.crosstab(df['Sleep_Hours_Non_Schoolnight'], df['Gender'])
print("Tabla Cruzada: Horas de Sueño (No clases) vs. Género:")
print(crosstab)

# a) ¿Qué porcentaje de estudiantes de la muestra son mujeres?
total_students = len(df)
num_female = len(df[df['Gender'] == 'Female'])
perc_female = num_female / total_students * 100
print(f"\na) Porcentaje de estudiantes que son mujeres: {perc_female:.2f}%")

# b) ¿Cuántas horas de sueño en promedio tienen los estudiantes? 😴
avg_sleep = df['Sleep_Hours_Non_Schoolnight'].mean()
print(f"\nb) Horas de sueño en promedio de todos los estudiantes: {avg_sleep:.2f} horas")

# c) ¿En términos generales, qué género duerme más tiempo?
avg_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].mean()
print("\nc) Promedio de horas de sueño por género:")
print(avg_sleep_by_gender)
# El género con mayor promedio indica quién duerme más.

# d) ¿Qué género tiene tiempos de sueño más homogéneos? (menor dispersión)
std_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].std()
print("\nd) Desviación estándar de las horas de sueño por género:")
print(std_sleep_by_gender)
# El género con menor desviación estándar tiene mayor homogeneidad.

# e) ¿Qué género presenta una mayor simetría en los tiempos de sueño? (skewness)
skew_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].skew()
print("\ne) Skewness de las horas de sueño por género:")
print(skew_sleep_by_gender)
# El género con un skewness más cercano a 0 presenta mayor simetría.

# f) ¿Qué género tiene un histograma más puntiagudo? ¿Y cuál uno más aplanado?
#     Se puede evaluar mediante la kurtosis: mayor kurtosis indica una distribución más puntiaguda.
kurtosis_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].apply(lambda x: x.kurtosis())
print("\nf) Kurtosis (apuntamiento) de las horas de sueño por género:")
print(kurtosis_sleep_by_gender)
# Un valor mayor de kurtosis indica un histograma más puntiagudo, 
# mientras que un valor menor (o negativo) indica un histograma más aplanado.

# 4. Opcional: Graficar histogramas por género para visualizar la distribución de las horas de sueño
plt.figure(figsize=(10, 6))
for gender in df['Gender'].unique():
    subset = df[df['Gender'] == gender]
    plt.hist(subset['Sleep_Hours_Non_Schoolnight'], bins=10, alpha=0.5, label=gender)
plt.title("Histograma de Horas de Sueño (No clases) por Género")
plt.xlabel("Horas de Sueño")
plt.ylabel("Número de Estudiantes")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
ct = pd.crosstab(df["Favorite_School_Subject"], df["Gender"])
print("Tabla Cruzada: Materia Preferida en el Colegio vs. Género")
print(ct)

# a) ¿A cuántos estudiantes hombres les gusta la historia? 📚
# Asumimos que el valor para Historia es "History".
num_history_men = ct.loc["History", "Male"] if ("History" in ct.index and "Male" in ct.columns) else 0
print(f"\na) Estudiantes hombres que prefieren historia: {num_history_men}")

# b) ¿Qué porcentaje de estudiantes son hombres?
total_students = len(df)
num_men = len(df[df["Gender"] == "Male"])
perc_men = (num_men / total_students) * 100
print(f"\nb) Porcentaje de estudiantes que son hombres: {perc_men:.2f}%")

# c) ¿Qué porcentaje de estudiantes prefieren música? 🎵
# Asumimos que el valor para Música es "Music".
num_music = df[df["Favorite_School_Subject"]=="Music"].shape[0]
perc_music = (num_music / total_students) * 100
print(f"\nc) Porcentaje de estudiantes que prefieren música: {perc_music:.2f}%")

# d) ¿De las mujeres, qué porcentaje prefieren otras actividades?
# Asumimos que el valor para Otras actividades es "Other activities".
female_total = len(df[df["Gender"]=="Female"])
female_other = df[(df["Gender"]=="Female") & (df["Favorite_School_Subject"]=="Other activities")].shape[0]
perc_female_other = (female_other / female_total) * 100 if female_total > 0 else 0
print(f"\nd) Porcentaje de mujeres que prefieren otras actividades: {perc_female_other:.2f}%")

# e) ¿De los estudiantes que prefieren Ciencias, qué porcentaje son hombres?
# Asumimos que el valor para Ciencias es "Sciences".
sciences_total = df[df["Favorite_School_Subject"]=="Sciences"].shape[0]
sciences_men = df[(df["Favorite_School_Subject"]=="Sciences") & (df["Gender"]=="Male")].shape[0]
perc_sciences_men = (sciences_men / sciences_total) * 100 if sciences_total > 0 else 0
print(f"\ne) Porcentaje de estudiantes que prefieren Ciencias y son hombres: {perc_sciences_men:.2f}%")

# f) ¿Qué porcentaje son mujeres y prefieren otras actividades?
perc_female_other_overall = (female_other / total_students) * 100
print(f"\nf) Porcentaje de estudiantes que son mujeres y prefieren otras actividades: {perc_female_other_overall:.2f}%")

# g) ¿Qué porcentaje son mujeres o prefieren música?
women_or_music = df[(df["Gender"]=="Female") | (df["Favorite_School_Subject"]=="Music")].shape[0]
perc_women_or_music = (women_or_music / total_students) * 100
print(f"\ng) Porcentaje de estudiantes que son mujeres o prefieren música: {perc_women_or_music:.2f}%")

# h) ¿De las mujeres, qué porcentaje prefiere música o ciencias?
female_music_or_sciences = df[(df["Gender"]=="Female") & (df["Favorite_School_Subject"].isin(["Music", "Sciences"]))].shape[0]
perc_female_music_or_sciences = (female_music_or_sciences / female_total) * 100 if female_total > 0 else 0
print(f"\nh) Porcentaje de mujeres que prefieren música o ciencias: {perc_female_music_or_sciences:.2f}%")
```

11b. Patrones de estudio: Ciencias vs. Artes
(Usamos "Favorite_School_Subject" para separar y "Doing_Homework_Hours" como proxy del patrón de estudio)

```python
df['Doing_Homework_Hours'] = pd.to_numeric(df['Doing_Homework_Hours'], errors='coerce')
science_students = df[df['Favorite_School_Subject'] == 'Science']
art_students = df[df['Favorite_School_Subject'] == 'Art']

# Opcional: Filtrar outliers en las horas de tarea (por ejemplo, descartar valores > 100 horas)
science_students = science_students[science_students['Doing_Homework_Hours'] <= 100]
art_students = art_students[art_students['Doing_Homework_Hours'] <= 100]

# Calcular el promedio de horas dedicadas a tareas para cada grupo
avg_homework_science = science_students['Doing_Homework_Hours'].dropna().mean()
avg_homework_art = art_students['Doing_Homework_Hours'].dropna().mean()

print("11b) Promedio de horas de tarea (proxy de patrón de estudio):")
print(f" - Estudiantes que prefieren Ciencia: {avg_homework_science:.2f} horas")
print(f" - Estudiantes que prefieren Arte: {avg_homework_art:.2f} horas")
```
