   - **a)** Â¿QuÃ© porcentaje de los estudiantes trabajan mÃ¡ximo 13 horas a la semana con pago? â±ï¸ğŸ”
   - **b)** Â¿QuÃ© porcentaje de los estudiantes trabajan mÃ­nimo 3 horas a la semana con pago? â±ï¸ğŸ”»
   - **c)** Â¿El 20% de los estudiantes trabajan mÃ­nimo cuÃ¡ntas horas con pago? â³
   - **d)** Â¿El 20% de los estudiantes trabajan mÃ¡ximo cuÃ¡ntas horas con pago? â³
   - **e)** Â¿En quÃ© intervalo se halla la moda? ğŸ“Š
   - **f)** Â¿El 50% de los estudiantes trabajan mÃ¡ximo cuÃ¡ntas horas con pago? ğŸ¯
   - **g)** Â¿En promedio, cuÃ¡ntas horas trabajan con pago a la semana? Â¿Y cuÃ¡l es la desviaciÃ³n estÃ¡ndar? ğŸ“ˆğŸ“‰
   - **h)** Â¿QuÃ© tipo de asimetrÃ­a presenta esta variable? ğŸ”„

```python
df['Paid_Work_Hours'] = pd.to_numeric(df['Paid_Work_Hours'], errors='coerce')
df = df[(df['Paid_Work_Hours'] >= 0) & (df['Paid_Work_Hours'] <= 40)]
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(df['Paid_Work_Hours'], bins=10, color='teal', edgecolor='black')
plt.title("Histograma: Tiempo de Trabajo Remunerado (horas)")
plt.xlabel("Horas de Trabajo")
plt.ylabel("NÃºmero de Estudiantes")
plt.tight_layout()
plt.show()

# a) Porcentaje de estudiantes que trabajan mÃ¡ximo 13 horas
perc_max_13 = (df['Paid_Work_Hours'] <= 13).mean() * 100

# b) Porcentaje de estudiantes que trabajan mÃ­nimo 3 horas
perc_min_3 = (df['Paid_Work_Hours'] >= 3).mean() * 100

# c) El 20% superior: valor a partir del cual el 20% de los estudiantes trabajan mÃ¡s horas
min_hours_top20 = df['Paid_Work_Hours'].quantile(0.80)

# d) El 20% inferior: valor por debajo del cual se encuentra el 20% de los estudiantes
max_hours_bottom20 = df['Paid_Work_Hours'].quantile(0.20)

# e) Intervalo donde se halla la moda: se selecciona el bin con mayor frecuencia
max_bin_index = np.argmax(counts)
mode_interval = (bins[max_bin_index], bins[max_bin_index + 1])

# f) La mediana (50% de los estudiantes trabajan hasta esa cantidad de horas)
median_hours = df['Paid_Work_Hours'].median()

# g) Promedio y desviaciÃ³n estÃ¡ndar
mean_hours = df['Paid_Work_Hours'].mean()
std_hours = df['Paid_Work_Hours'].std()

# h) AsimetrÃ­a: calcular el coeficiente de skewness
skewness = df['Paid_Work_Hours'].skew()
if skewness > 0:
    skew_type = "asimetrÃ­a positiva (sesgada a la derecha)"
elif skewness < 0:
    skew_type = "asimetrÃ­a negativa (sesgada a la izquierda)"
else:
    skew_type = "simÃ©trica"

print(f"a) Porcentaje que trabajan mÃ¡ximo 13 horas: {perc_max_13:.2f}%")
print(f"b) Porcentaje que trabajan mÃ­nimo 3 horas: {perc_min_3:.2f}%")
print(f"c) El 20% superior trabaja al menos: {min_hours_top20:.2f} horas")
print(f"d) El 20% inferior trabaja como mÃ¡ximo: {max_hours_bottom20:.2f} horas")
print(f"e) La moda se encuentra en el intervalo: {mode_interval}")
print(f"f) La mediana (50% de los estudiantes) trabaja como mÃ¡ximo: {median_hours:.2f} horas")
print(f"g) Promedio: {mean_hours:.2f} horas, DesviaciÃ³n estÃ¡ndar: {std_hours:.2f}")
print(f"h) Tipo de asimetrÃ­a: {skew_type} (skewness = {skewness:.2f})")
```

```python
df['Work_At_Home_Hours'] = pd.to_numeric(df['Work_At_Home_Hours'], errors='coerce')
df = df[(df['Work_At_Home_Hours'] >= 0) & (df['Work_At_Home_Hours'] <= 60)]
num_bins = 10  # Ajusta el nÃºmero de bins segÃºn lo desees
counts, bin_edges = np.histogram(df['Work_At_Home_Hours'], bins=num_bins)
cum_counts = np.cumsum(counts)
total = len(df['Work_At_Home_Hours'].dropna())
cum_relative = cum_counts / total
plt.figure(figsize=(10, 6))
plt.plot(bin_edges[1:], cum_relative, marker='o', color='blue', drawstyle='steps-post')
plt.title("Ojiva: Tiempo Dedicado a Quehaceres del Hogar")
plt.xlabel("Horas de Ayuda en el Hogar")
plt.ylabel("Frecuencia Acumulada (proporciÃ³n)")
plt.grid(True)
plt.tight_layout()
plt.show()

# a) Porcentaje de estudiantes que dedican mÃ­nimo 15 horas
perc_min_15 = (df['Work_At_Home_Hours'] >= 15).mean() * 100

# b) Porcentaje de estudiantes que dedican mÃ¡ximo 5 horas
perc_max_5 = (df['Work_At_Home_Hours'] <= 5).mean() * 100

# c) MÃ¡ximo nÃºmero de horas que el 15% de los estudiantes ayudan en sus casas
#    Se interpreta como el valor para el percentil 0.15 (el 15% inferior).
max_hours_15percent = df['Work_At_Home_Hours'].quantile(0.15)

# d) Porcentaje de estudiantes que ayudan al menos 5 horas
perc_at_least_5 = (df['Work_At_Home_Hours'] >= 5).mean() * 100

# e) MÃ­nimo nÃºmero de horas que el 35% de los estudiantes ayudan
#    Se interpreta como el percentil 0.35
min_hours_35percent = df['Work_At_Home_Hours'].quantile(0.35)

# f) Intervalo donde se halla la moda
#    Se asume que la "moda" en un histograma es el bin con mayor frecuencia.
max_bin_index = np.argmax(counts)
mode_interval = (bin_edges[max_bin_index], bin_edges[max_bin_index+1])

# g) Promedio y desviaciÃ³n estÃ¡ndar
mean_hours = df['Work_At_Home_Hours'].mean()
std_hours = df['Work_At_Home_Hours'].std()

# h) Tipo de asimetrÃ­a (skewness)
skewness = df['Work_At_Home_Hours'].skew()
if skewness > 0:
    skew_type = "asimetrÃ­a positiva (sesgada a la derecha)"
elif skewness < 0:
    skew_type = "asimetrÃ­a negativa (sesgada a la izquierda)"
else:
    skew_type = "distribuciÃ³n simÃ©trica"

print("RESULTADOS:\n")
print(f"a) Porcentaje que dedican mÃ­nimo 15 horas: {perc_min_15:.2f}%")
print(f"b) Porcentaje que dedican mÃ¡ximo 5 horas: {perc_max_5:.2f}%")
print(f"c) MÃ¡ximo nÃºmero de horas (15% de los estudiantes): {max_hours_15percent:.2f}")
print(f"d) Porcentaje que ayudan al menos 5 horas: {perc_at_least_5:.2f}%")
print(f"e) MÃ­nimo nÃºmero de horas (35% de los estudiantes): {min_hours_35percent:.2f}")
print(f"f) Intervalo de la moda (bin mÃ¡s frecuente): {mode_interval}")
print(f"g) Promedio: {mean_hours:.2f} horas, DesviaciÃ³n estÃ¡ndar: {std_hours:.2f}")
print(f"h) Tipo de asimetrÃ­a: {skew_type} (skewness = {skewness:.2f})")
```

9. **Tabla Cruzada: Horas de SueÃ±o vs. GÃ©nero**  
   Construye una tabla cruzada entre la variable â€œÂ¿CuÃ¡ntas horas de sueÃ±o por noche suele dormir cuando no tiene clases al dÃ­a siguiente?â€ y â€œGÃ©neroâ€. Responde:
   - **a)** Â¿QuÃ© porcentaje de estudiantes de la muestra son mujeres?
   - **b)** Â¿CuÃ¡ntas horas de sueÃ±o en promedio tienen los estudiantes? ğŸ˜´
   - **c)** Â¿En tÃ©rminos generales, quÃ© gÃ©nero duerme mÃ¡s tiempo?
   - **d)** Â¿QuÃ© gÃ©nero tiene tiempos de sueÃ±o mÃ¡s homogÃ©neos?
   - **e)** Â¿QuÃ© gÃ©nero presenta una mayor simetrÃ­a en los tiempos de sueÃ±o?
   - **f)** Â¿QuÃ© gÃ©nero tiene un histograma mÃ¡s puntiagudo? Â¿Y cuÃ¡l uno mÃ¡s aplanado? ğŸ“‰ğŸ“Š

```python
df['Sleep_Hours_Non_Schoolnight'] = pd.to_numeric(df['Sleep_Hours_Non_Schoolnight'], errors='coerce')
crosstab = pd.crosstab(df['Sleep_Hours_Non_Schoolnight'], df['Gender'])
print("Tabla Cruzada: Horas de SueÃ±o (No clases) vs. GÃ©nero:")
print(crosstab)

# a) Â¿QuÃ© porcentaje de estudiantes de la muestra son mujeres?
total_students = len(df)
num_female = len(df[df['Gender'] == 'Female'])
perc_female = num_female / total_students * 100
print(f"\na) Porcentaje de estudiantes que son mujeres: {perc_female:.2f}%")

# b) Â¿CuÃ¡ntas horas de sueÃ±o en promedio tienen los estudiantes? ğŸ˜´
avg_sleep = df['Sleep_Hours_Non_Schoolnight'].mean()
print(f"\nb) Horas de sueÃ±o en promedio de todos los estudiantes: {avg_sleep:.2f} horas")

# c) Â¿En tÃ©rminos generales, quÃ© gÃ©nero duerme mÃ¡s tiempo?
avg_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].mean()
print("\nc) Promedio de horas de sueÃ±o por gÃ©nero:")
print(avg_sleep_by_gender)
# El gÃ©nero con mayor promedio indica quiÃ©n duerme mÃ¡s.

# d) Â¿QuÃ© gÃ©nero tiene tiempos de sueÃ±o mÃ¡s homogÃ©neos? (menor dispersiÃ³n)
std_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].std()
print("\nd) DesviaciÃ³n estÃ¡ndar de las horas de sueÃ±o por gÃ©nero:")
print(std_sleep_by_gender)
# El gÃ©nero con menor desviaciÃ³n estÃ¡ndar tiene mayor homogeneidad.

# e) Â¿QuÃ© gÃ©nero presenta una mayor simetrÃ­a en los tiempos de sueÃ±o? (skewness)
skew_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].skew()
print("\ne) Skewness de las horas de sueÃ±o por gÃ©nero:")
print(skew_sleep_by_gender)
# El gÃ©nero con un skewness mÃ¡s cercano a 0 presenta mayor simetrÃ­a.

# f) Â¿QuÃ© gÃ©nero tiene un histograma mÃ¡s puntiagudo? Â¿Y cuÃ¡l uno mÃ¡s aplanado?
#     Se puede evaluar mediante la kurtosis: mayor kurtosis indica una distribuciÃ³n mÃ¡s puntiaguda.
kurtosis_sleep_by_gender = df.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].apply(lambda x: x.kurtosis())
print("\nf) Kurtosis (apuntamiento) de las horas de sueÃ±o por gÃ©nero:")
print(kurtosis_sleep_by_gender)
# Un valor mayor de kurtosis indica un histograma mÃ¡s puntiagudo, 
# mientras que un valor menor (o negativo) indica un histograma mÃ¡s aplanado.

# 4. Opcional: Graficar histogramas por gÃ©nero para visualizar la distribuciÃ³n de las horas de sueÃ±o
plt.figure(figsize=(10, 6))
for gender in df['Gender'].unique():
    subset = df[df['Gender'] == gender]
    plt.hist(subset['Sleep_Hours_Non_Schoolnight'], bins=10, alpha=0.5, label=gender)
plt.title("Histograma de Horas de SueÃ±o (No clases) por GÃ©nero")
plt.xlabel("Horas de SueÃ±o")
plt.ylabel("NÃºmero de Estudiantes")
plt.legend()
plt.tight_layout()
plt.show()
```
