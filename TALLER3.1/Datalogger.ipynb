```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path of the file to read
csv_path = r"C:\Users\cesar\Desktop\platzi\Quick_programming_tasks\TALLER3.1\datalogger2.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_path)
```

```python
print(df.info())
```

```python
# Convert the 'DATE' column to datetime
df['DATETIME'] = pd.to_datetime(df['DATE'] + ' ' + df['TIME'])
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
df['SN'] = df['SN'].astype(int)
```

```python
print(df.head())
```

```python
df = df.drop(columns=['DATE', 'TIME'])
```

```python
print(df.head)
```

```python
print(df.head())
```

```python
df['DATETIME'] = pd.to_datetime(df['DATETIME'])
```

```python
print(df.info())
```

```python
df.rename(columns={'\toC': 'T[C]'}, inplace=True)
```

```python
df.rename(columns={'\t%RH': 'RH[%]'}, inplace=True)
```

```python
print(df.head())
```

```python
df = df.drop(columns=['\tDP'])
```

```python
print(df.head())
```

```python
df['T[C]'] = pd.to_numeric(df['T[C]'], errors='coerce')
df = df[(df['T[C]'] >= 0) & (df['T[C]'] <= 50)]
plt.plot(df['DATETIME'], df['T[C]'])
plt.xlabel('DATETIME')
plt.xticks(rotation=45)
plt.ylabel('T[C]')
plt.title('Grfico de T[C] vs DATETIME')
plt.show()

```

```python
df['RH[%]'] = pd.to_numeric(df['RH[%]'], errors='coerce')
df = df[(df['RH[%]'] >= 0) & (df['RH[%]'] <= 100)]
plt.plot(df['DATETIME'], df['RH[%]'])
plt.xlabel('DATETIME')
plt.xticks(rotation=45)
plt.ylabel('T[C]')
plt.title('Grfico de RH[%] vs DATETIME')
plt.show()
```

```python
max_row = df.loc[df['T[C]'].idxmax()]
print("el dia en el la temperatura fue mayor fue")
print(max_row)
min_row = df.loc[df['T[C]'].idxmin()]
print("el dia en el la temperatura fue menor fue") 
print(min_row)
max_row = df.loc[df['RH[%]'].idxmax()]
print("el dia en el la humedad fue mayor fue")
print(max_row)
min_row = df.loc[df['RH[%]'].idxmin()]
print("el dia en el la humedad fue menor fue")
print(min_row)
```
